## Model

```python
num = pd.concat([num,df['isFraud']],axis = 1)
```

```python
num.head(2)
```

```python
num['diff_trans_open'] = (pd.to_datetime(num['accountOpenDate'])-pd.to_datetime(num['transactionDate'])).dt.days
```

```python
num['diff_currexp_last_addr'] = (pd.to_datetime(num['currentExpDate'])-pd.to_datetime(num['dateOfLastAddressChange'])).dt.days
```

```python
num.drop(['currentExpDate','accountOpenDate','dateOfLastAddressChange','transactionDate'],axis = 1,inplace = True)
```

```python
num.head(3)
```

```python
# cat.head(3)
```

```python
dmy = pd.get_dummies(cat, drop_first=True)
final_df200 = pd.concat([dmy,num], axis = 1)
```

```python
num.head()
```

```python
final_df200.shape
```

```python
final_df200.replace({True: 1, False: 0}, inplace=True)
```

```python
X = final_df200.drop('isFraud',axis = 1)
y = final_df200['isFraud']
xc = sm.add_constant(X)
X_train,X_test,y_train,y_test = train_test_split(xc, y, test_size=0.2,random_state=1)
```

# KNN Code

```python
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
```

```python
ypred_knn_train = knn_model.predict(X_train)
ypred_knn_test = knn_model.predict(X_test)
```

```python
accuracy_score(y_train,ypred_knn_train)
```

```python
accuracy_score(y_test,ypred_knn_test)
```

```python
confusion_matrix(y_test, ypred_knn_test)
```
